This is a tutorial of Matplotlib.



```python
# Import Libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# create  a figure and axes
fig, ax = plt.subplots()

# Plot data on the axes
ax.plot([1, 2, 3], [9, 10, 11])
ax.set_title('Simple Line Plot')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')

plt.show()
```

Line Plot

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x, y, marker='o', linestyle='--', color='r', label='Line Plot')
plt.legend()
plt.show()
```

Scatter Plot
ideal for two variables

```python
x = [1, 2, 3, 4, 5, 9, 12, 11]
y = [10, 20, 25, 30, 90, 40, 45, 60]
plt.scatter(x, y, color='r', label='Scatter Plot')
plt.legend()
plt.show()
```

Bar plot
can be h -  horizontally or v - vertically

```python
categories = ['A', 'B', 'C', 'D']
values = [7, 13, 20, 10]
plt.bar(categories, values, color='g', label='Bar Plot')
plt.legend()
plt.show()
```

Histogram
used for distribution of data

```python
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
plt.hist(data, bins=4, color='purple', label='Histogram')
plt.legend()
plt.show()
```

Pie Chart
Used for propotions

```python
sizes = [30, 40, 20, 10]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels,autopct='%1.1f%%')
plt.show()
```

```python
sizes = [30, 40, 20, 10]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels,autopct='%1.0f%%')
plt.show()
```

autopct='%1.0f%%' states the decimal points

# Customize plot and Titles

```python
plt.plot([1, 2, 3], [4, 5, 6])
plt.title('Customize Plot', fontsize=18, fontweight='bold')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

Customize the legend

```python
plt.plot([1, 2, 3], [4, 5, 6], label='Line 1')
plt.plot([1, 2, 3], [6, 5, 4], label='Line 2')
plt.legend()
plt.show()
```

Customize grides

```python
plt.plot([1, 2, 3], [4, 5, 6])
plt.grid(True) # Add grid
plt.show()
```

# Subplots
create mutiple plots in one frame

```python
fig, ax = plt.subplots(2,2) #2 rows, 2 columns
ax[0,0].plot([1, 2, 3], [4, 5, 6])
ax[0,1].scatter([1, 2, 3], [4, 5, 6])
ax[1,0].bar([1, 2, 3], [4, 5, 6])
ax[1,1].hist([1, 2, 2, 3, 3, 3, 4, 4, 4, 4])
plt.show()
```

```python
# Themes and Styles
plt.style.use('ggplot')
plt.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

```python
# Saving the plot
plt.plot([1, 2, 3], [4, 5, 6])
plt.savefig('my_plot.png', dpi = 300) # correct funtion name
plt.show()
```

# Sample Dataset
charts and analysis to create
- Describe your data: Shape, missing values, duplicated values and data types
- Create a line chart to view the trend.
- Create a bar chart to view the number of passangers per month
- create a scatter to view the relationship between two variables of your choice 

```python
# Load the dataset
flight= sns.load_dataset('flights')

# view the dataset
flight
```

```python
# shape of the dataset
flight.shape
```

```python
# check for missing values
flight.isnull().sum()
```

```python
# check for dulicate values
flight.duplicated().sum()
```

```python
# display the information of the dataset
flight.info()
```

```python
# Display the summary statistics of the dataset
flight.describe()
```

```python
# Create a line chart to view the trends
plt.figure(figsize=(10, 6))
sns.lineplot(x='year', y='passengers', data=flight)
plt.title('Passenger Traffic Over the Years')
```

```python
# Create a bar chat to view the passenger per month
categories = flight['month']
values = flight['passengers']
plt.bar(categories, values, color='g', label='Bar Plot')
plt.legend()
plt.show()
```

```python
# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='year', y='passengers', data=flight)
plt.title('Passengers over the years')
plt.show()
```
