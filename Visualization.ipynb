```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
# display first five rows
df.head()

df.shape
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
# remove duplicates
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
# Usually the first step in data analysis
df.describe()
```

Bar chart - Number of Suvivors by class

```python
# Count the number of surivors
class_count = df.groupby('class')["survived"].sum()

# Create the bar chart
plt.figure(figsize=(10,5))
plt.bar(class_count.index,class_count.values, color = ['blue', 'green', 'red'])

# Customizing the chart
plt.xlabel("Class")
plt.ylabel("Number of Survivors")
plt.title("Survivors by Passenger Class")

#show the plot
plt.show()
```

Histograms

Distributio of Passanger Ages

```python
plt.figure(figsize=(6,4))
plt.hist(df["age"].dropna(),
         bins = 10,
         color= 'skyblue',edgecolor='black')
plt.title('Passanger Age Distribution')
```

```python
plt.figure(figsize=(6,4))
plt.hist(df["age"].dropna(),
         bins = 20,
         color= 'blue',edgecolor='black')
plt.title('Passanger Age Distribution')
```

Fare Distribution of Passengers

```python
plt.figure(figsize=(10,5))
sns.histplot(df["fare"],bins=30, kde=True, color='purple')
plt.xlabel("Fare")
plt.ylabel("Count")
plt.title("Fare Distribution of Titanic Passengers")
plt.show()
```

Violin Plot - Age and Gender Distribution

```python
plt.figure(figsize=(10,5))
sns.violinplot(x="sex", y="age", data=df, palette='muted', split=True)
plt.xlabel("Gender")
plt.ylabel("Age")
plt.title("Age Distribution by Gender")
plt.show()
```

```python
# Violin plot by class
plt.figure(figsize=(10,5))
sns.violinplot(x="class", y="age", data=df, palette='muted', split=True)
plt.xlabel("Gender")
plt.ylabel("Age")
plt.title("Age Distribution by class")
plt.show()
```

```python
# Violin plot by fare
plt.figure(figsize=(10,5))
sns.violinplot(x="sex", y="fare", data=df, palette='muted', split=True)
plt.xlabel("Gender")
plt.ylabel("Age")
plt.title("Fare Distribution by Gender")
plt.show()
```

Subplots - Survival Rate by Sex

```python
fig, axes = plt.subplots(1, 2, figsize=(10,4))

# Bar plot for survival by gender
sns.barplot(x="sex", y="survived", data=df, ax=axes[0], palette='coolwarm')
axes[0].set_title("Survival Rate by Gender")

# Pie chart for overall survival distribution
df["survived"].value_counts().plot.pie(autopct ="%1.1f%%",labels=["Did Not survive", "Survived"], colors=["lightcoral", "lightgreen"], ax=axes[1])
axes[1].set_title("Overall Survival Distribution")
plt.tight_layout()
plt.show()
```

Scatter Plot - Age vs. Fare

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x="age", y="fare", hue="survived", data=df, palette="coolwarm",
                alpha=0.6)
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Fare vs Age (Color by Survival)")
plt.legend(title="Survived")
plt.show()
```

```python
# Select numerical columns
numerical_df = df.select_dtypes(include=['float64', 'int64'])

# Create a correlation matrix
corr_matrix = numerical_df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of Numerical Variables in Titanic Dataset')
plt.show()
```
