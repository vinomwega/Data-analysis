Data Cleaning

```python
# Drop the first row (if it's a header or unnecessary)
sdg = sdg_data.iloc[1:]

# Drop columns with all missing values
sdg  = sdg.dropna(axis=1, how='all')

# Drop columns containing 'ID' or 'Id' in their name (case-insensitive)
to_drop = [col for col in sdg.columns if 'id' in col.lower()[-2:]]
sdg  = sdg.drop(to_drop, axis=1)

# Drop columns containing 'order' in their name
to_drop = [col for col in sdg.columns if 'order' in col.lower()]
sdg  = sdg.drop(to_drop, axis=1)

# Drop columns that contain only a single unique value (not useful for analysis)
unique_columns = sdg.loc[:, sdg.nunique() == 1].columns
sdg  = sdg.drop(unique_columns, axis=1)

# Drop specific column 'CharacteristicLabel' if it exists in the dataset
sdg = sdg.drop(['CharacteristicLabel'], axis=1)

# Convert the 'Value' column to float 
sdg['Value'] = sdg['Value'].astype(float)
```
