```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# 3d scatter plot
np.random.seed(42)
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

# create 3d figure
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

#scatter plot
ax.scatter(x, y, z, c=z, cmap='coolwarm', marker='o')

#labels and title
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title('3D Scatter Plot')

plt.show()
```

```python
# Create 2d scatter plot
plt.figure(figsize=(6,5))
plt.scatter(x, y, c=z, cmap='coolwarm', edgecolors='k')

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('2D Scatter Plot(Color Represents Z axis)')
plt.colorbar(label='Z Axis')

plt.show()
```
