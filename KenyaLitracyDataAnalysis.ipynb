```python
#drop First row
#df.drop([0], inplace=True)
df =  data.iloc[1:]

# Drop columns with null values
#df = df.dropna(axis=0, how='all') # drop rows that all have null values
df = df.dropna(axis=1, how='all') # drop columns that all have null values

# Drop Id Columns
to_drop = [column for column in df.columns if 'Id' in column or'ID' in column] # this is list comprehension where runs a for loop to check for id in columns to be deleted
df = df.drop(to_drop, axis=1)# drops all rows that have id in them

# Drop order column
to_drop = [column for column in df.columns if 'Order' in column ] 
df = df.drop(to_drop, axis=1)

#Drop duplicate column
df = df.drop(['CharacteristicLabel'], axis=1)

# Drop Columns with single value
to_drop =  df.loc[:, df.nunique() == 1]. columns
df = df.drop(to_drop, axis=1)

# Column type conversion
df['Value'] = df['Value'].astype(float)


# view new data set
df.head()
```

Questions

Gender Disparity: In which countries is the gap btn men and women's litracy the largest

```python
# slice on indicator

#df['Indicator']. unique() # shows the values in Indicator
literacy_ind = [ind for ind in df['Indicator']. unique()if 'literate' in ind]
#literacy_ind
gender_gap = df[df['Indicator'].isin(literacy_ind)]

# slice on county data
gender_gap = gender_gap [gender_gap['LevelRank'] == 2]

#Pivot our data
gender_gap = gender_gap.pivot_table(index= 'Location', columns='Indicator', values='Value')

# Calculate difference for men and women
gender_gap['gap'] = abs(gender_gap.iloc[:,0] - gender_gap.iloc[:,1])

# Sort values
gender_gap = gender_gap.sort_values('gap', ascending=False)

# view Gender disparity data
gender_gap.sample(5)
```
