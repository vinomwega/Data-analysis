```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
#3d scatter plot
np.random.seed(42)
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

#create 3d figure
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

#scatter plot
ax.scatter(x,y,z,c=z,cmap='coolwarm',marker='o')

#labels and title
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title('3D Scatter Plot')

plt.show()
```

```python
#create a 2d scatter plot
plt.figure(figsize=(8,6))
plt.scatter(x,y,c=z,cmap='coolwarm',edgecolors='k')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('2D Scatter Plot(color represents Z)')
plt.colorbar(label='Z Value')

plt.show()
```

```python
#creating a 3d surface plot
#define the 3d function
X= np.linspace(-5,5,50)
Y= np.linspace(-5,5,50)
X,Y = np.meshgrid(X,Y)
Z= np.sin(np.sqrt(X**2 + Y**2))

#create 3d figure
fig =plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

#surface plot
ax.plot_surface(X,Y,Z, cmap='viridis')

#labels and title
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title('3D Surface Plot')

plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8,6))
sns.heatmap(Z, cmap='viridis', xticklabels=False, yticklabels=False)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('2D Heatmap(From 3D surface Data)')
plt.show()
```

```python
#3D Contour plot
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

#contour plot
ax.contour3D(X, Y, Z, 50, cmap='plasma')

ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('X Axis')
ax.set_title('3D contour Plot')

plt. show()
```
