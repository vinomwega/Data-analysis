```python
from pywaffle import Waffle
import matplotlib.pyplot as plt
```

```python
# Sample Data: Social Media usage share
social_media_usage = {
    "Facebook": 40,
    "Instagram": 25,
    "Twitter": 25,
    "TikTok": 10,
    "Snapchat": 10
}

# Define brand colors for each
brand_colors = {
    "Facebook": "#3b5998", # Facebook blue
    "Instagram": "#C13584", # Instagram pink
    "Twitter": "#1DA1F2", #Twitter Blue
    "Snapchat": "#FFFC00", # Snapchat Yellow
    "TikTok": "#25F4EE" # Tiktok teal
}

# Create Waffle
fig = plt.figure(
    FigureClass= Waffle,
    rows=5,
    values=social_media_usage,
    colors= [brand_colors[platform] for platform in social_media_usage.keys()],
    title={"label": "social_media_usage Distribution", "loc":"center", "fontsize":18},
    legend={
        'loc': 'lower center', # position legend, bottom
        'bbox_to_anchor': (0.5,-0.5), # adjust position
        'ncol': 3,# number of columns in legend
        'fontsize': 10,
        'frameon': False # remove legend boder
    }
)
# Adjust layout to make space for the legend
plt.tight_layout()
plt.show()
```

```python
from wordcloud import WordCloud
```

```python
# Sample text data
text =""" Facebook Instagram Twiiter  Snapchat Tiktok Instagram Instagram Tiktok Tiktok snapchat"""

# Generate word cloud
wordcloud = WordCloud(width=800, height= 400, background_color='white').generate(text)

# Display the word cloud
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off') # Turn off axis
plt.title('Word Cloud of social Media Mentions')
plt.show()
```

```python
# Sample Text
reviews = [
    " I love Using Tiktok since it is my facorite app",
    "I enjoy Facebook alot since it is qiute vast",
    "I love using Instagram it more fun than other apps",
    "Snapchat allows for you to take fun photoes that you cant get anywhere else"
]
# combine the list into a single string
text = ' '.join(reviews)

#genate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Display the word cloud
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off') # Hides the axis
plt.title("Word Cloud pf Survey Text")
plt.show()
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download("stopwords")

from sklearn.datasets import fetch_20newsgroups # gets the exact dataset you will need
```

```python
# Using news articles
#Load dataser
news = fetch_20newsgroups(subset='all')
text_data = " ".join(news.data[:200]) # Taking the first 200

# Generate Word cloud
wordcloud= WordCloud(
    width=800, height=400, background_color="black",
).generate(text_data)

# Display
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud of News Articles", fontsize=14)
plt.show()
```

```python
# Get English stopwords from NLTK
stop_words = set(stopwords.words('english'))

# Generate word cloud with stopwords removed
wordcloud = WordCloud(
    width=800, height=400, background_color="black", stopwords=stop_words #Explicitly removing stopwords
).generate(text_data)

# Display
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud of News Articles(Stopwords removed)", fontsize=14)
plt.show()
```
