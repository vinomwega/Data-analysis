# Multiple Line regression
y = B0 + B1 * x1 + B2 * x2 + B3 * x3 + E

Where :
- y is the dependent variable (the variable we are trying to predict) = House Price
- B0 is the intercept or constant term
- B1, B2, B3 are the coefficients of the independent variables
- E is the error term (the difference between the observed and predicted values)
- x1 = House Age
- x2 = Distance of MRT Satation
- x3 = No of convience stores


```python
X = data[["house age", "distance to the nearest MRT station", "number of convenience stores"]]
X = sm.add_constant(X)  # Add constant term for regression
y = data["house price of unit area"]

# Multiple Linear Regression
model = sm.OLS(y, X).fit()
print("\nMultiple Linear Regression Results: Predicting Average Score")
print(model.summary())
```

# Overall Analysis
* **Adjusted R-squared(0.538)** - states that 53.8% of the variance of the house prices depend on the independent variables(house age, distance to the nearest MRT station, number of convenience stores)
* **Goodness of fit(5.44e-69)** - states that the it is significantly lower than the p - value of 0.05 hence it is staistically significant here by indicating that at least one of the independant variables can be used to calculate the price of the house.
* **Regression equation:**  y = 42.9773 + -0.2529(x1) + -0.0054(x2) + 1.2974(x3) + 0

    * **const(42.9773)** - this shows the value o the house if all other variables are zero
    * **house age (-0.2529)** - this shows that the age  of the house is negatively correlated with the price of the house and statistically significant. This means that the older the house, the lower the price. Therefore decreasing by -25.29% per year.
    * **distance to the nearest MRT station(-0.0054)** - this shows that the distance to the nearest MRT station is negatively correlated with the price of the house and statistically significant. This means that the further away the house is from the nearest MRT station, the lower the price. A decrease by -0.54% per unit of distance.
    * **number of convenience (1.2974)** - this shows that the number of convience stores positively correlated with the price of the house and statistically significant. This means that the more convenience stores there are near the house the higher the price of the house. An increase of 1.30 times more per convenience store.
    * **E** (0) - Assume there was no other external factor
